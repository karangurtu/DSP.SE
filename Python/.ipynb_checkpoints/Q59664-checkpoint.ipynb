```python
from numpy import random
from scipy.linalg import toeplitz, inv
from pylab import figure, clf, plot, xlabel, ylabel, xlim, ylim, title, grid, axes, show


N = 5
h = [1,1,-1,1,1]

x = random.normal(0, 0.01, N)
X = toeplitz(x)
H = toeplitz(h)
y = H @ x
h_hat = inv(X.transpose() @ X) @ X.transpose() @ y

figure(1)
plot(h)
title("True FIR filter")

figure(2)
plot(y)
plot(x,'r.')
title("Input (red) and output (blue) of filter")

figure(3)
plot(h)
plot(h_hat,'r.')
```
